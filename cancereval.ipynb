# Creating models

```python
#select features and the target
y=data['diagnosis'] 
X=data.drop('diagnosis', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder

#Scale all te features so they will be treated equally
scaler=StandardScaler()
X=scaler.fit_transform(X)

#Convert the target into binary for easier classification
ohe=OneHotEncoder(drop='first', sparse_output=False)
y=pd.get_dummies(y, drop_first=True, dtype=int)
y_ohe = pd.DataFrame(ohe.fit_transform(y))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer

#create a train and test split
X_train, X_test, y_train, y_test = train_test_split(X,y_ohe,test_size=0.3, random_state=1)

#impute the missing values
imputer = SimpleImputer(strategy='mean') 
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```

```python
#create a logistic model

#import neccesary library
from sklearn.linear_model import LogisticRegression

log = LogisticRegression(random_state=1)
log.fit(X_train, y_train)

#make a prediction for the model
y_pred_train = log.predict(X_train)
y_pred_test = log.predict(X_test)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

#let's calculate the precision, recall, accuracy and f1_score for the training data

#calculate the precision
logreg_train_precision = precision_score(y_train,y_pred_train)
#clculate the recall
logreg_train_recall = recall_score(y_train,y_pred_train)
#calculate the accuracy
logreg_train_accuracy = accuracy_score(y_train,y_pred_train)
#calculate the f1_score
logreg_train_f1_score = f1_score(y_train,y_pred_train)
```

```python
#let's calculate the precision, recall, accuracy and f1_score for the test data

from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

#calculate the precision
logreg_precision = precision_score(y_test, y_pred_test)
#clculate the recall
logreg_recall = recall_score(y_test, y_pred_test)
#calculate the accuracy
logreg_accuracy = accuracy_score(y_test,y_pred_test)
#calculate the f1_score
logreg_f1_score = f1_score(y_test,y_pred_test)
```

```python
#Compare the results for the training and test data
print('The precision for this training model is: ', logreg_train_precision)
print('The precision for the test model is: ', logreg_precision)
print('\n\n')


print('The recall for this training model is: ', logreg_train_recall)
print('The recall for the test model is: ', logreg_recall)
print('\n\n')

print('The accuracy of this training model is: ', logreg_train_accuracy)
print('The accuracy of the test model is: ', logreg_accuracy)
print('\n\n')

print('The f1_score of this training model is: ', logreg_train_f1_score)
print('The f1_score of the test model is: ', logreg_f1_score)
```

```python
from sklearn.metrics import classification_report
class_report_train = classification_report(y_train, y_pred_train)
class_report_test = classification_report(y_test, y_pred_test)


print('The classification report for the training model is:','\n', class_report_train)
print('\n')
print('The classification report for the test model is :','\n', class_report_test)
```

# Decision tree model

## Using a CART tree model


```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=42, max_depth=3)
regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score

# Make predictions and evaluate 
y_train_pred = regressor.predict(X_train)
y_test_pred = regressor.predict(X_test)
print(' Training MSE score:', mse(y_train, y_train_pred))
print('Training R-sq score:', r2_score(y_train,y_train_pred))
print('\n')
print(' Test MSE score:', mse(y_test, y_test_pred))
print(' Test R-sq score:', r2_score(y_test,y_test_pred))
```

### hyper parameter tuning

```python
from sklearn.metrics import r2_score, mean_squared_error

def performance(y_true, y_predict):

    # Calculate the r2 score between 'y_true' and 'y_predict'
    r2 = r2_score(y_true, y_predict)

    # Calculate the root mean squared error between 'y_true' and 'y_predict'
    rmse = mean_squared_error(y_true, y_predict, squared=False)

    # Return the score
    return [r2, rmse]
```

```python
# Identify the optimal tree depth for given data
import matplotlib.pyplot as plt
%matplotlib inline
max_depths = list(range(1, 31))
mse_results = []
r2_results = []

for max_depth in max_depths:
    regressor = DecisionTreeRegressor(max_depth=max_depth, random_state=45)
    regressor.fit(X_train, y_train)
    y_pred = regressor.predict(X_test)
    score = performance(y_test, y_pred)
    r2_results.append(score[0])
    mse_results.append(score[1])

plt.figure(figsize=(12, 6))
plt.plot(max_depths, r2_results, 'b', label='R2')
plt.xlabel('Tree Depth')
plt.ylabel('R-squared')
plt.legend()
plt.show()
plt.figure(figsize=(12, 6))
plt.plot(max_depths, mse_results, 'r', label='RMSE')
plt.xlabel('Tree Depth')
plt.ylabel('RMSE')
plt.legend()
plt.show()
```

```python
# Identify the optimal minimum split size for given data
import numpy as np
min_samples_splits = np.arange(2, 11)
mse_results = []
r2_results = []

for min_samples_split in min_samples_splits:
    regressor = DecisionTreeRegressor(min_samples_split=int(min_samples_split),random_state=45)
    regressor.fit(X_train, y_train)
    y_pred = regressor.predict(X_test)
    score = performance(y_test, y_pred)
    r2_results.append(score[0])
    mse_results.append(score[1])

plt.figure(figsize=(12, 6))
plt.plot(min_samples_splits, r2_results, 'b', label='R2')
plt.show()
plt.figure(figsize=(12, 6))
plt.plot(min_samples_splits, mse_results, 'r', label='RMSE')
plt.show()
```
